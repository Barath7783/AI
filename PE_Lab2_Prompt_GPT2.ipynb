```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
# Load GPT-2 base model
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
# IMPORTANT FIX: set pad_token to eos_token
tokenizer.pad_token = tokenizer.eos_token
model.config.pad_token_id = tokenizer.eos_token_id
```

```python
# Final Prompt created using 5 principles
prompt = """Write a simple, clear paragraph explaining the future of artificial intelligence in education.Keep the explanation between 80–100 words.Present the paragraph in plain English suitable for college students."""
```

```python
# Encode prompt
#inputs = tokenizer.encode(prompt, return_tensors="pt")
# Tokenize with attention mask
inputs = tokenizer(
    prompt,
    return_tensors="pt",
    padding=True,
    truncation=True
)
```

```python
# Generate response
outputs = model.generate(
    input_ids=inputs["input_ids"],
    attention_mask=inputs["attention_mask"],   # FIX 1
    max_length=250,
    temperature=0.7,
    do_sample=True,
    top_p=0.9,
    pad_token_id=tokenizer.eos_token_id          # FIX 2
)
```

```python
# Decode output
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```
